```python
# input text
with open(os.path.join(os.path.dirname(os.path.abspath("__file__")) ,'.\profile.json')) as profile_json:
    # seeking profule insights
    profile = personality_insights.profile(
        profile_json.read(),
        'application/json',
        content_type='application/json',
        consumption_preferences=True,
        raw_scores=True
    ).get_result()

# display results
print(json.dumps(profile, indent=2))
```
